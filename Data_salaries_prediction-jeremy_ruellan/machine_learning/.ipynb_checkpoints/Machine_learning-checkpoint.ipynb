# train_test_split

```python
X = df_final2.drop(['Salary', 'Company', 'Description'], axis=1)
X
```

```python
y = df_final2.Salary
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
```

# Régressions

## Régression linéaire multiple

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train, y_train)
y_pred = lm.predict(X_test)
```

```python
plt.figure(figsize=(12,12))
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='red', linewidth=3)
plt.xlabel("Prix")
plt.ylabel("Prediction de prix")
plt.title("Prix reels vs predictions")
```

```python
sns.distplot(y_test-y_pred)
```

```python
#RMSE
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
#R Squared (R2)
scoreR2 = r2_score(y_test, y_pred)
print(scoreR2)
```

```python
lm.score(X_test,y_test)
```

## Régression par forêts aléatoires

```python
from sklearn import ensemble

rf = ensemble.RandomForestRegressor()
rf.fit(X_train, y_train)
y_rf = rf.predict(X_test)
print(rf.score(X_test,y_test))
```

```python
plt.figure(figsize=(12,12))
plt.scatter(y_test, y_rf)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='red', linewidth=3)
plt.xlabel("Prix")
plt.ylabel("Prediction de prix")
plt.title("Prix reels vs predictions")
```

```python
sns.distplot(y_test-y_rf)
```

```python
#RMSE
print(np.sqrt(mean_squared_error(y_test, y_rf)))
```

```python
#R Squared (R2)
r2 = r2_score(y_test, y_rf)
print(r2)
```

```python
rf.score(X_test,y_test)
```

## Régression par XGBoost

```python
import xgboost as xgb

regressor=xgb.XGBRegressor(eval_metric='rmsle')
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {"max_depth":    [4, 5],
              "n_estimators": [500, 600, 700],
              "learning_rate": [0.01, 0.015]}
```

```python
search = GridSearchCV(regressor, param_grid, cv=5).fit(X_train, y_train)

print("The best hyperparameters are ",search.best_params_)
```

```python
regressor=xgb.XGBRegressor(learning_rate = search.best_params_["learning_rate"],
                           n_estimators  = search.best_params_["n_estimators"],
                           max_depth     = search.best_params_["max_depth"],
                           eval_metric='rmsle')

regressor.fit(X_train, y_train)
```

```python
predictions = regressor.predict(X_test)
```

```python
plt.figure(figsize=(12,12))
plt.scatter(y_test, predictions)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='red', linewidth=3)
plt.xlabel("Prix")
plt.ylabel("Prediction de prix")
plt.title("Prix reels vs predictions")
```

```python
sns.distplot(y_test-predictions)
```

```python
#RMSE
print(np.sqrt(mean_squared_error(y_test, predictions)))
```

```python
#R Squared (R2)
r2 = r2_score(y_test, predictions)
print(r2)
```
